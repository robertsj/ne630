## Is a Resonance Narrow or Wide?

# The Self-Shielding Effect

Consider the 6.67 eV resonance and the NR spectrum with 10-1, 100-1, and 1000-1 mixtures of H-1 and U-238.

```python
sigma_t_10 = cross_section_data["U238_t"]  + 1*20
sigma_t_100 = cross_section_data["U238_t"] + 100*20 
sigma_t_1000 = cross_section_data["U238_t"] + 1000*20 

plt.semilogy(E, 100/(E*sigma_t_10), lw=1, label="$\phi_{NR}(E)$, 10-1")
plt.semilogy(E, 1000/(E*sigma_t_100), lw=1, label="$\phi_{NR}(E)$, 100-1")
plt.semilogy(E, 10000/(E*sigma_t_1000), lw=1, label="$\phi_{NR}(E)$, 1000-1")

plt.semilogy(E,  cross_section_data["U238_t"], lw=1, label="$\sigma_t(E)$")

plt.xlim(6, 7.5)
plt.ylim(1e-3, 1e4)
plt.xlabel("$E$ [eV]"); plt.legend(loc=1);
```
