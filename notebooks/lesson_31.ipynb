# The Diffusion Approximation

Our solution is to apply **Fick's Law**, which states

$$
  \vec{J}(\vec{r}) = -D(\vec{r}) \vec{\nabla} \phi(\vec{r})
  \tag{6.11}
$$

This is a *phenomenological* model, i.e., not first principles, but it just "works."



(From *Fundamentals of Heat and Mass Transfer, 6e*, which some of you may know!)

The diffusion coefficient:

$$
  D = \frac{1}{3\Sigma_{tr}} = \frac{1}{3(\Sigma_{t} - \bar{\mu}\Sigma_s)}
$$

The average cosine of the scattering angle (if scattering is isotropic in the CM system):

$$
  \bar{\mu} = \frac{2}{3A}
$$

**Question**: What does this say about $\Sigma_{tr}$ compared to $\Sigma_{t}$?

Now, substitution of Fick's law into our continuity equation gives

$$
 - \vec{\nabla} \cdot D(\vec{r}) \vec{\nabla} \phi(\vec{r})
  + \Sigma_a(\vec{r}) \phi(\vec{r}) = s'''(\vec{r})
  + \nu\Sigma_f(\vec{r}) \phi(\vec{r}) \, ,
  \tag{6.12}
$$

which is the one-group (or one-speed or monoenergetic) **neutron diffusion equation**.

$$
\begin{split}
\frac{1}{v} \frac{\partial \phi}{\partial t}
 &- \nabla D(\mathbf{r}, E) \nabla \phi 
  + \Sigma_t(\mathbf{r}, E) \phi(\mathbf{r}, E, t) \\
  &= \int^{\infty}_0 \Sigma_s(\mathbf{r}, E'\to E) \phi(\mathbf{r}, E', t) dE' \\
  &+ \frac{\chi(E)}{k_{\text{eff}}} \int^{\infty}_0 \bar{\nu}(\mathbf{r}, E')\Sigma_f(\mathbf{r}, E') \phi(\mathbf{r}, E', t) dE' + S(\mathbf{r}, E, t) \, .
\end{split}
$$
