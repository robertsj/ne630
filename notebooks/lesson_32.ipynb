```python
import numpy as np
import matplotlib.pyplot as plt

def s(x, a, b):
    return 1 + b*np.cos(np.pi*x/(2*a))

def phi(x, a, D, SigmaA, b):
    L = np.sqrt(D/SigmaA)
    C1 = -L**2 * np.exp(a/L) / (D*(np.exp(2*a/L)+1))
    C2 = C1
    C3 = L**2/D
    C4 = (b/D) / (1/L**2 + (np.pi/(2*a))**2)
    return C1*np.exp(x/L) + C2*np.exp(-x/L) + C3 + C4*np.cos(np.pi*x/(2*a))
```

```python
a = 10
x = np.linspace(-a, a, 1000)
b = -1
D = 1.0
SigmaA = 0.1
```

```python
plt.figure(figsize=(10, 4))
plt.plot(x, phi(x, a, D, SigmaA, b))
plt.plot(x, s(x, a, b))
```
