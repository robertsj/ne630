```python
import numpy as np
import matplotlib.pyplot as plt
def phi(x, a, D, a, s0):
    L = np.sqrt(D/a)
    C1 = -s0*L**2/D*(1 + (np.exp(a/L)-1)/(np.exp(-a/L)-np.exp(a/L)))
    C2 = s0*L**2/D* (np.exp(a/L)-1)/(np.exp(-a/L)-np.exp(a/L))
    return C1*np.exp(x/L) + C2*np.exp(-x/L) + s0*L**2/D

s0 = 1.0
D  = 1.0
a = 0.1
x = np.linspace(0, 1, 1000)
plt.figure(figsize=(10, 4))
a=10;  plt.plot(x, phi(a*x, a, D, a, s0), 'k-',  label=f"a={a}")
a=20;  plt.plot(x, phi(a*x, a, D, a, s0), 'k-.', label=f"a={a}")
a=100; plt.plot(x, phi(a*x, a, D, a, s0), 'k--', label=f"a={a}")
plt.xlabel('$x/a$')
plt.ylabel('$\phi(x)$')
plt.legend();
plt.savefig("flux_plot.pdf")
```
