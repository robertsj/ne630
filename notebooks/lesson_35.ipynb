# Criticality

For $k_{\infty} > 1$, we found

$$ 
   \phi(x) = \frac{s'''}{D \alpha^{2}} \left(\frac{\cos{\left(\alpha x \right)}}{\cos{\left(\frac{a \alpha}{2} \right)}} - 1\right )
$$

where 

$$
  \alpha^2 = \frac{k_{\infty}-1}{L^2} = \frac{\nu\Sigma_f/\Sigma_a-1}{D/\Sigma_a} \equiv B^2_m = \text{material buckling}
$$


We also noted that when $a\alpha/2 = \pi/2$, the term $\cos(a\alpha/2)$ vanishes, and the flux blows up:

$$
\begin{split}
  B_m^2 \equiv \frac{k_{\infty}-1}{L^2} &= \left ( \frac{\pi}{a} \right )^2 \equiv B_g^2 = \text{geometric buckling}\\
 \rightarrow k_{\infty} &= L^2 \left ( \frac{\pi}{a} \right )^2 + 1 \\
 \rightarrow 1          &= \frac{k_{\infty}}{ 1 + L^2 \left ( \frac{\pi}{a} \right )^2  } = k_{\infty} P_{NL} = \frac{\text{gains}}{\text{losses from absorption}} \times \frac{\text{losses from absorption}}{\text{losses}} \\
           &= \frac{ \nu\Sigma_f }{ \Sigma_a + D \left ( \frac{\pi}{a} \right )^2  } = \frac{\text{gains}}{\text{losses}} \, .
\end{split}  
$$

Each of these statements corresponds to *criticality*, i.e., when losses are exactly balances by gains:

$$
   {\large \boxed{k = k_{\infty} P_{NL} =   1}} \, .
$$

Just like a real critical system, we get a steady state flux only if there is no source!

But what happens when $B_m^2 \neq B_g^2$? As described in Section 7.2, we can still characterize the steady state flux (i.e., solve the NDE) by introducing $k$ as a new *parameter*.  For slab geometry, that looks like 

$$
  -D \frac{d^2\phi}{dx^2} + \Sigma_a \phi(x) = \frac{\nu\Sigma_f}{k} \phi(x) \, .
$$

or 

$$
 \frac{d^2\phi}{dx^2} + \overbrace{\left ( \frac{k_{\infty}/k-1}{L^2} \right )}^{B^2} \phi(x)  = 0\, .
 \tag{Lewis (7.4), but for slabs}
$$

In other words, $k$ reduces or increases $k_{\infty}$ (or, as suggested in the text, $\nu$) so that $B_m^2 = B_g^2$.  Our goal, then, is either (1) to determine $k$ given the materials and geometry or (2) to determine the materials or geometry that leads to a given $k$ (usually 1!).  

## Worked Example: The Critical Buckling 

<div class="alert alert-block alert-success">
Consider the same system illustrated above, but now let's apply vacuum conditions (the book discusses the simpler case of setting $\phi = 0$ at the boundaries; see Eqs. (7.15-7.17)).
</div>

<div class="alert alert-block alert-info">

**Step 1**:  Our equations are

$$
  \phi'' + B^2 \phi(x) = 0\, , \qquad J^+(-a/2) = 0\, , \quad \text{and} \qquad J^-(a/2) = 0 \, .
$$
    
</div>

<div class="alert alert-block alert-info">

**Step 2**:  The general solution is

$$
  \phi(x) = C_1 \cos(B x) + C_2 \sin(B x) \, .
$$

</div>

<div class="alert alert-block alert-info">


**Step 3**:  Apply the boundary conditions:

$$
\begin{split}
  J^+(-a/2) &= \frac{\phi(-a/2)}{4} - \frac{D}{2}\frac{d\phi}{dx}\Big|_{x=-a/2} \\
    &= \frac{C_{1} \cos{\left(\frac{B a}{2} \right)}}{4} - \frac{C_{2} \sin{\left(\frac{B a}{2} \right)}}{4} - \frac{D \left(B C_{1} \sin{\left(\frac{B a}{2} \right)} + B C_{2} \cos{\left(\frac{B a}{2} \right)}\right)}{2} \\
    &= 0
\end{split}
    \tag{left BC}
$$


$$
\begin{split}
    J^-(a/2) 
    &= \frac{\phi(a/2)}{4} + \frac{D}{2}\frac{d\phi}{dx}\Big|_{x=a/2} \\
    &= \frac{C_{1} \cos{\left(\frac{B a}{2} \right)}}{4} + \frac{C_{2} \sin{\left(\frac{B a}{2} \right)}}{4} + \frac{D \left(- B C_{1} \sin{\left(\frac{B a}{2} \right)} + B C_{2} \cos{\left(\frac{B a}{2} \right)}\right)}{2} \\
    &= 0
\end{split}
    \tag{right BC}
$$
    
</div>

<div class="alert alert-block alert-info">

Subtraction of the first from the second leads to

$$
    \displaystyle \frac{C_{2} \cdot \left(2 B D \cos{\left(\frac{B a}{2} \right)} + \sin{\left(\frac{B a}{2} \right)}\right)}{2} = 0 \longrightarrow C_2 = 0 \, .
$$

and substitution of $C_2 = 0$ into our left boundary condition leads to 

$$
\frac{C_{1} \left(- 2 B D \sin{\left(\frac{B a}{2} \right)} + \cos{\left(\frac{B a}{2} \right)}\right)}{4} = 0
$$
    
</div>

<div class="alert alert-block alert-info">

**Step 4**: This is where our process departs from everything else we've done: we can't "solve for" $C_1$, and the equation is satisfied only for the right value(s) of $B$.  This is the nature of *eigenvalue* problems.  Simplifying our expression a bit leads to 

$$
 \boxed{ \cot{\frac{Ba}{2}} = 2BD } \, .
 \tag{criticality condition}
$$

The critical buckling $B^2$ is the *smallest positive value* that satisfies this criticality condition and corresponds to a strictly positive $\phi(x)$ called the **fundamental mode** and the **largest eigenvalue** $k$ (i.e., the multiplication factor).  I repeat: smallest $B^2$, biggest $k$.
    
</div>

<div class="alert alert-block alert-info">

Without numbers, this is about as far as we can go! For comparison, recall that the zero-flux conditions give rise to a much simpler definition for the critical buckling: $B^2 = \pi^2/a^2$ (with $(3\pi)^2/a^2,  (5\pi)^2/a^2$, etc., also satisfying the criticality condition but not the positive flux condition).
    
</div>

## Beyond Buckling: $k$ and $\phi$

Don't forgot that the buckling isn't the final answer!  Once we have $B^2$, we have

$$
 k = \frac{k_{\infty}}{1+L^2 B^2} \, ,
$$

which we can evaluate given $k_{\infty}$ (or $\nu\Sigma_f$), $\Sigma_a$, and $D$.  

## Worked Example: The Critical Eigenvalue

<div class="alert alert-block alert-success">


In **Problem [35.1]**, you should have found that $B \approx 0.06$ or $B^2 \approx 0.0036$.  For $D=1$ cm, $\Sigma_a = 0.1$ 1/cm, and $k_{\infty} = 1.5$, determine the critical eigenvalue $k$.

</div>

<div class="alert alert-block alert-info">

**Solution**: We have

$$
   k = \frac{k_{\infty}}{1 + L^2 B^2} = \frac{1.5}{1 + (1/0.1)0.0036} = 1.4479 \, .
$$

In other words, this reactor is *supercritical*.  In order to make this system critical, we need to decrease $k_{\infty}$ (e.g., by reducing enrichment) or increasing leakage (e.g., by reduces $a$ which increases $B^2$).
    
</div>

Given $B^2$, we also have

$$
   \phi(x) = C \cos(Bx) \, .
$$

Here, $C = C_1$ is our remaining constant of integration, and, like all *eigenfunctions*, its magnitude is arbitrary without some normalization constraint.  

For a critical reactor, that constraint comes from the average reactor power density $P''' = \gamma \Sigma_f \bar{\phi}$ [J/cm$^3$], where $\bar{\phi}$ is the spatially-averaged flux, or

$$
   P''' = \gamma \Sigma_f C_1 \frac{1}{a} \int^{a/2}_{-a/2} \cos(Bx) dx 
   \longrightarrow \boxed{ C_1 = \frac{a B P'''}{2 \gamma \Sigma_f \sin(Ba/2)}} \, .
$$



For a power reactor (with finite dimensions), the total power $P$ is the integral of $P'''$ over the core volume, and, for a finite, cylindrical core, Eq. (7.29) provides a similar expression for $C$.

# The NDE in Cylindrical Coordinates

The NDE in 1-D (radial) cylindrical coordinates is 

$$
   \frac{1}{r}\frac{d}{dr}r\frac{d\phi}{dr} + B^2 \phi(r) = 
      \boxed{\frac{d^2 \phi}{dr^2} + \frac{1}{r} \frac{d\phi}{dr} + B^2 \phi(r) 
      = 0} \, , \qquad r \ge 0 \, .
$$

We'll apply a boundary condition at some finite value of $r$ and require that the flux is bounded everywhere (i.e., less than $\infty$).

The "problem" with that boxed equation is that the coefficient of $\phi'$ is $1/r$, and that *isn't constant!*

## Back to the Basics (of Math)

Recall that a polite$^*$ function $f(x)$ can be written as

$$
    f(x) = f(0) + f'(0)x + \frac{f''(0) x^2}{2}  + \frac{f'''(0) x^3}{6} + \ldots = \sum^{\infty}_{i=0} a_i x^i \, .
$$

<font size=2pt>(*kidding, sort of; "polite" as in "well behaved" as in the function is continuous, its derivatives exist, it blows up gently, and all the other niceties we like in a function.)</font>

One way to solve an ODE is to assume its solution is such an infinite sum of $x^i$.  Let's test the waters with our more familiar NDE in slab geometry (without multiplication).  If we substitute 

$$
  \phi(x) = \sum^{\infty}_{i=0} a_i x^i
$$

into 

$$
  -\frac{d^2\phi}{dx^2} + \frac{1}{L^2} \phi(x) = 0 \, 
$$

we get

$$
 -\frac{d^2}{dx^2} [a_0 + a_1 x + a_2 x^2 + a_3 x^3 + a_4 x^4 \ldots ] + \frac{1}{L^2} [a_0 + a_1 x + a_2 x^2 + \ldots] = 0 
$$

$$
 - [(2)(1)a_2 + (3)(2) a_3 x + (4)(3) a_4 x^2 \ldots ] + \frac{1}{L^2} [a_0 + a_1 x + a_2 x^2 + \ldots] = 0 
$$

For this to be satisfied for any value of $x$, then each power of $x$ must be balanced:

$$
\begin{split}
    (2)(1) a_2  &= \frac{1}{L^2} a_0  \\
    (3)(2) a_3  &= \frac{1}{L^2} a_1  \\
    (4)(3) a_4  &= \frac{1}{L^2} a_2  \\
\end{split}
$$

and so on...

...and a pattern emerges:

$$
   a_2 = \frac{L^{-2} a_0}{(2)(1)}
$$

$$
   a_4 = \frac{L^{-2} a_2}{(4)(3)} =  \frac{L^{-2}}{(4)(3)}\frac{L^{-2} a_0}{(2)(1)} = 
    \frac{L^{-4} a_0}{4!}
$$

With a dash of fairy magic, we get

$$
   a_{2i} = \frac{L^{-2i} a_0}{(2i)!} \quad \longrightarrow \quad \boxed{\phi_{\text{even}}(x) = a_0 \sum^{\infty}_{i=0} \frac{(x/L)^{2i}}{(2i)!}= a_0 \cosh(x/L)}  \, .
$$

...and the odd terms yield $a_1 \sinh(x/L)$.

You might recall that $e^x = 1 + x + x^2/2 + x^3/6 + \ldots$.  Appendix A also gives

$$
  \cosh(x) = \frac{e^{x}+e^{-x}}{2} \qquad \text{and} \qquad \sinh(x) = \frac{e^{x}-e^{-x}}{2} \, .
$$

Thus, $ae^{x} + be^{-x}$ can be written as $(a+b)\cosh(x) + (a-b)\sinh(x)$.  

```python
import matplotlib.pyplot as plt
import numpy as np
from math import factorial
def phi_even(x):
    v = x**0
    for i in range(1, 25): # for i too large, the factorial is too big to evaluate!
        v += x**(2*i)/float(factorial(2*i))
    return v
x = np.linspace(-1, 1, 50)
plt.figure(figsize=(8,3))
plt.plot(x, phi_even(x), 'k', label="$\phi_{\mathrm{even}}(x)$")
plt.plot(x, np.cosh(x), 'r*', label="$\cosh(x)$")

plt.xlabel("$x$")
plt.legend();
```

Now, back to cylinders and solutions of

$$
\frac{d^2 \phi}{dr^2} + \frac{1}{r} \frac{d\phi}{dr} + B^2 \phi(r) 
      = 0
$$

Substitutition of $\phi(r) = \sum^{\infty}_{i=0} a_i r^i$ into this equation leads to

$$
[(2)(1) a_2 + (3)(2) a_3 r + (4)(3) a_4 r^2 + \ldots] 
  + \frac{1}{r} [a_1 + 2 a_2 r + 3 a_3 r^2 + \ldots ]
 + B^2 [a_0 + a_1 r + \ldots] = 0 \, .
$$

Now, the $1/r$ could mean things misbehave at $r=0$, but let's be daring and assume we can multiply through by $r$:

$$
[(2)(1) a_2 r + (3)(2) a_3 r^2 + (4)(3) a_4 r^3 + \ldots] 
  + [a_1 + 2 a_2 r + 3 a_3 r^2 + \ldots ]
 + B^2 [a_0 r + a_1 r^2 + \ldots] = 0 \, .
$$


Matching like powers yields

$$
\begin{split}
   a_1 &= 0 \\
   (2)(1) a_2 + 2 a_2 + B^2 a_0 &= 0  \\
   (3)(2) a_3 + 3 a_3 + B^2 a_1 &= 0 \\
   (4)(3) a_4 + 4 a_4 + B^2 a_2 &= 0 \\
\end{split}
$$

and so on.  

The odd terms are apparently all zero, but the even terms form recurrence relationship similar to the one above for $\cosh(x/L)$, and the result is the **Bessel function of the first kind of order 0**:

$$
  J_0(Br) = \sum^{\infty}_{i=0} (-1)^i \frac{ \left ( \frac{(Br)^2}{4} \right )^i }{(i!)^2} \, .
$$

(see https://mathworld.wolfram.com/BesselFunctionoftheFirstKind.html)

When needed, you can evaluate $J_0(x)$ using `scipy.special.jv(0, x)`. For example:

```python
from scipy.special import jv 
r = np.linspace(0, 5)
plt.figure(figsize=(8,3))
plt.plot(r, jv(0, r))
plt.xlabel('$r$');
```

We can also use SymPy:

```python
import sympy as sy
r = sy.symbols("r")
sy.besselj(0, r)
```

Were we to require the gradient (for current continuity, vacuum conditions, etc.), we'd need the derivative of $J_0(r)$:

```python
sy.diff(sy.besselj(0, r), r)
```

This relationship $J'_0(r) = -J_1(r)$ and several other properties of Bessel functions are listed in Appendix B.
